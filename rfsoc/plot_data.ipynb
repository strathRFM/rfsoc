```python
# libraries
import numpy as np
import matplotlib.pyplot as plt
import os

```

```python
# load the spectrum file

fname = "testspectrum.txt"

with open(fname) as f:
    data = f.readlines()
    data = [float(i) for i in data]
    
N = len(data)
DF = 4
fs = 4096
cf = 1024

#lower_lim = cf-((fs/DF)/2)
#upper_lim = cf+((fs/DF)/2)
lower_lim = 0
upper_lim = 2048
step = (upper_lim - lower_lim)/N

print("data points (fft size): "+ str(N))
print("upper: "+ str(upper_lim) + " MHz")
print("lower: "+ str(lower_lim) + " MHz")
print("step: "+ str(step) + " MHz (resolution)")

step = (upper_lim - lower_lim)/N
db = np.transpose(np.arange(lower_lim,upper_lim,step))

# plot the spectrum using matplotlib

print("plotting")
plt.plot(db,data)
plt.axis([lower_lim,upper_lim,-140, 0])
plt.ylabel("dBFS")
plt.xlabel("frequency [MHz]")
plt.show()
```
